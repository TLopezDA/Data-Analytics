```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df.head(10)
```

```python
#Check that the values under "SchoolDBN" are unique (*Unable to remove index column because values under "SchoolDBN" are not unique
df["School DBN"].is_unique
```

```python
#View columns
df.columns
```

```python
#remove columns
drop_columns = ['Number Scoring CR', 'Percent Scoring CR']
df.drop(drop_columns, inplace = True, axis = 1)
```

```python
df.head(5)
```

```python
#Simple Visualizations
df.groupby(['School Name'])['Year'].count().sort_values(ascending=True)
```

```python
#formating 4 digit years/convert 'Year' to string
regex = r'^(\d{4})'
df['Year'].astype(str).str.extract(regex, expand=False)
df.head()
```

```python
#Data Wrangling - basic transformations - subset columns
df_subset = df[['School DBN', 'School Name', 'School Level', 'Regents Exam', 'Year', 'Mean Score', 'Number Scoring 80 or Above', 'Percent Scoring 80 or Above']]
df_subset.head(5)
```

```python
#Cleaning task - Remove rows with "s" values
df_subset = df_subset[df_subset['Mean Score'] != 's']
df_subset.head()
```

```python
#View all schools in the Bronx
bronx=df_subset[df_subset['School DBN'].str.contains('08X',na=False)]
bronx.head()
```

```python
#Pick a feature/measure/score - Bronx Schools w/ Mean Score >80. Create 'bx80' df.
#Convert 'Mean Score' column to numeric
df_subset.loc['Mean Score'] = pd.to_numeric(df_subset.loc['Mean Score'])
bx80 = df_subset[(df_subset['Mean Score'] >80)&(df_subset['School DBN'].str.contains('08X'))]
bx80.head()
```

```python
#View schools listed in the bx80 df.
bx80.groupby(['School Name'])['Year'].count().sort_values(ascending=True)
```

```python
#Pick any school as comparison point - Is DeWitt Clinton HS listed? Create 'clinton' df
clinton=df_subset[df_subset['School Name'].str.contains('DeWitt Clinton High School', na=False)]
clinton
```

```python
# Compare one school to schools in a particular borough

import matplotlib.pyplot as plt

# Filter the bronx dataframe to include only a specific Regents Exam (e.g., Common Core Algebra)
bronx_algebra = bronx[bronx['Regents Exam'] == 'Common Core Algebra']

# Filter the clinton dataframe to include only a specific Regents Exam (e.g., Common Core Algebra)
clinton_algebra = clinton[clinton['Regents Exam'] == 'Common Core Algebra']

# Group by School Name and calculate the average Mean Score for each school in the Bronx
average_scores_bronx = bronx_algebra.groupby('School Name')['Mean Score'].mean()

# Calculate the average Mean Score for DeWitt Clinton High School
average_score_clinton = clinton_algebra['Mean Score'].mean()

# Compare DeWitt Clinton High School's average score to the average scores of other Bronx schools
print(f"Average Mean Score for DeWitt Clinton High School: {average_score_clinton}")
print("Average Mean Scores for other Bronx schools:")
print(average_scores_bronx)

# Bar chart visualization
plt.figure(figsize=(10, 6))
plt.bar(average_scores_bronx.index, average_scores_bronx.values, label='Bronx Schools', color='g')
plt.axhline(y=average_score_clinton, color='r', linestyle='-', label='DeWitt Clinton High School')
plt.xlabel('School Name', color='g')
plt.ylabel('Average Mean Score', color='g')
plt.title('Comparison of Common Core Algebra Mean Scores between DeWitt Clinton High School and other Bronx Schools')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend()
plt.show()
```
