```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
df.head()
```

```python
#show all columns
df.columns
```

```python
#groupby race
df.groupby('race').size()
```

```python
#subset df columns
df_subset=df[['race', 'gender', 'age','time_in_hospital','num_lab_procedures', 'num_procedures', 'num_medications',
'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses','diabetesMed', 'readmitted']]
df_subset.head()
```

```python
#total male/female in 50-60 age bracket
df_subset[df_subset['age']=='[50-60)'].groupby(['gender']).size()
```

```python
#view data for only age
grp=df_subset[df_subset('age')].head()
grp
```

```python
#Group1
grp1=df_subset[df_subset['gender']=='Female']
grp1
```

```python
#Group2
grp2=df_subset[df_subset['gender']=='Male']
grp2.head()
```

```python
#summary statistics for female subset
print(grp1.describe())
```

```python
#summary statistics for male subset
print(grp2.describe())
```

```python
#Look at the num_lab_procedures amongst the different races for females
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=grp1, x='num_lab_procedures', hue='race')
plt.show()
```

```python
#Look at the num_lab_procedures amongst the different races for males
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=grp2, x='num_lab_procedures', hue='race', bins=50)
plt.show()
```

```python
#Look at the num_lab_procedures amongst the different races for females
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=grp1, x='time_in_hospital', hue='race')
plt.show()
```

```python
#Look at the time_in_hospital amongst the different races for males
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=grp2, x='time_in_hospital', hue='race')
plt.show()
```

I created two subsets to look at the data for females and compare them to males. The summary statistics show that although the count is higher for females under each category, there is not much of a difference in the mean for each one. I then created plots to look at it from the perspective of race. I plotted both the time_in_hospital and num_lab_procedures and conclude that the the Caucasian race seeks medical attention at a much greater rate than the other races.
