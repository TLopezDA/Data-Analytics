```python
# Subset AGE_GROUP 25-44 for ARREST_PRECINCT 19 and 47
precincts = [19, 47]
subset_df = df[df['AGE_GROUP'] == '25-44'][df['ARREST_PRECINCT'].isin(precincts)]
```

```python
subset_df.head()
```

```python
#Plot Precincts 19 and 47 for specific offenses

import matplotlib.pyplot as plt
#Plot df_subset
offense = ['ASSAULT 3 & RELATED OFFENSES', 'FELONY ASSAULT', 'PETIT LARCENY', 'GRAND LARCENY', 'BURGLARY', 'ROBBERY', 'SEX CRIMES', 'DANGEROUS WEAPONS']
sns.histplot(data=subset_df[subset_df['OFNS_DESC'].isin(offense)], x='OFNS_DESC', hue='ARREST_PRECINCT')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plot line graph comparing 19 and 47 precincts for all offenses

import matplotlib.pyplot as plt

# Subset AGE_GROUP 25-44 for ARREST_PRECINCT 19 and 47
precincts = [19, 47]
subset_df = df[(df['AGE_GROUP'] == '25-44') & (df['ARREST_PRECINCT'].isin(precincts))]

# Group by precinct and offense, then count occurrences
offense_counts = subset_df.groupby(['ARREST_PRECINCT', 'OFNS_DESC']).size().reset_index(name='counts')

# Create the line plot
plt.figure(figsize=(12, 6))
for precinct in precincts:
    precinct_data = offense_counts[offense_counts['ARREST_PRECINCT'] == precinct]
    plt.plot(precinct_data['OFNS_DESC'], precinct_data['counts'], label=f'Precinct {precinct}', marker='o')

plt.xlabel('Offense Description')
plt.ylabel('Number of Arrests')
plt.title('Comparison of Offenses between Precincts 19 and 47 (Age Group 25-44)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

I selected the 47th Precinct in the Bronx - where I was born and raised - and compared it to the 19th Precinct in Manhattan.

The top offenses for both Precincts involve individuals in the 25-44 age group.

The top three races reported by the 47th Precinct are Black, White Hispanic and Asian/Pacific Islander.

The top three races reported by the 19th Precinct are Black, Black Hispanic and White Hispanic.

The top two offenses reported by the 47th Precinct are Assault and Felony Assault.

The top three offenses reported by the 19th Precinct are Petit Larceny and Grand Larceny.
